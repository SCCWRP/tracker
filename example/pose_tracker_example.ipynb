# Import the `Tracker` class from `pose`

```python
from __future__ import print_function
from pose import Tracker
```

#  Initialize the `Tracker` class

```python
print(Tracker.__doc__)
```

```python
tracker = Tracker(video_source = "/Users/Jake/Downloads/2015-07-01-1-v5.mp4")
#tracker = Tracker(0) # uncomment this to use webcam
```

# Show and record the video stream

```python
print(Tracker.show_video.__doc__)
```

```python
tracker.show_video(resize = 1, video_output = "/Users/Jake/Desktop/test.mp4")
```

# Show and save a still image from the stream

```python
print(Tracker.show_still.__doc__)
```

```python
tracker.show_still(frame_number = 200, figsize = (20,10), still_output = "/Users/Jake/Desktop/test.png")
```

# Set the region of interest

```python
print(Tracker.set_roi.__doc__)
```

```python
tracker.set_roi(pt1 = (250,5), pt2 = (1350,1080), imshow = True, figsize = (20,6))
```

# Set the tracking parameters

```python
print(Tracker.set_contour_area.__doc__)
```

```python
print(Tracker.set_background.__doc__)
```

```python
print(Tracker.set_fps.__doc__)
```

```python
print(Tracker.set_morphology.__doc__)
```

```python
print(Tracker.set_blur.__doc__)
```

```python
tracker.set_contour_area(10,10000)
tracker.set_background(1000,300)
tracker.set_fps(60)
tracker.set_morphology(10,10)
tracker.set_blur(11, 2)
```

# Start the tracker, save the data, and record the video output

```python
print(Tracker.track.__doc__)
```

```python
tracker.track("/Users/Jake/Desktop/data.csv", video_output = "/Users/Jake/Desktop/test.mp4", show_bg = True, show_mask = True, vtrail_len = 200)
```
